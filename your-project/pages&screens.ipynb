```python
def open_amazon_link(url):
    # Function to open the Amazon search link in the default web browser
    webbrowser.open(url)

def filter_items():
    user_choice = choice_var.get()
    user_description = description_entry.get()

    if user_choice.lower() == "movies":
        df = pd.read_csv("top_250_movies_by_genre.csv")
    elif user_choice.lower() == "books":
        df = pd.read_csv("books_df_with_description.csv")
    else:
        messagebox.showerror("Error", "Invalid choice")
        return

    df["overview"].fillna("", inplace=True)
    tokens = word_tokenize(user_description.lower())
    filtered_tokens = [token for token in tokens if token not in stopwords.words("english")]

    filtered_items = df[df["overview"].str.contains("|".join(filtered_tokens))]

    if not filtered_items.empty:
        top_items = filtered_items.sort_values(by="rating", ascending=False).head(5)

        result_window = Toplevel(window)
        result_window.title("Results")
        result_window.geometry("800x400")

        for index, row in top_items.iterrows():
            # Create a label for each result with title, description, and rating
            item_label = tk.Label(result_window, text=f"Title: {row['title']}\nDescription: {row['overview']}\nRating: {row['rating']}")
            item_label.pack()
            
            # Generate an Amazon search link based on the title
            amazon_search_url = f"https://www.amazon.es/s?k={user_choice}+{'+'.join(row['title'].split())}"
            
            # Create a button to open the Amazon search link
            open_button = tk.Button(result_window, text="Open Amazon Search", command=lambda url=amazon_search_url: open_amazon_link(url))
            open_button.pack()
    else:
        messagebox.showinfo("Results", f"No {user_choice.capitalize()} found matching the description.")

window = tk.Tk()
window.title("Filter")
window.geometry("800x400")

choice_label = tk.Label(window, text="Choose 'movies' or 'books':")
choice_label.pack()
choice_var = tk.StringVar()
choice_entry = tk.Entry(window, textvariable=choice_var)
choice_entry.pack()

description_label = tk.Label(window, text="Enter description:")
description_label.pack()
description_entry = tk.Entry(window)
description_entry.pack()

filter_button = tk.Button(window, text="Filter Items", command=filter_items)
filter_button.pack()

window.mainloop()
```
